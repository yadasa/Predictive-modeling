```python
#install the dependencies"
import numpy as np
import pandas as pd
import csv
from pathlib import Path
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
plt.style.use('bmh')
```

```python
csvPath = Path("Resources/BTC-USD.csv")
```

```python
#read the stock data
with open(csvPath, 'r') as csvFile:
    csvReader = csv.reader(csvFile)
```

```python
#create dataframe
dF = pd.read_csv(csvPath)
dF.head(7)
```

```python
# Get number of trading days in dataframe
dF.shape
```

```python
#Visualize the data for each daily closing price
plt.figure(figsize=(16,8))
plt.title("Bitcoin")
plt.xlabel("Days")
plt.ylabel("$ Price (USD)")
plt.plot(dF['Close'])
plt.show
```

```python
#Get the close price
dF = dF[['Close']]
dF.head(7)
```

```python
# Initialize variable that predicts x days into the future
futureDays = 28
#Creat new column (target) shifted x united up
dF['Prediction'] = dF[['Close']].shift(-futureDays)
dF.tail(7)
```

```python
# Create the future data set 'x' and convert it to Numpy Array, remove the last 'x' rows
x = np.array(dF.drop(['Prediction'], 1))[:-futureDays]
#print(x)
```

```python
# Create target data set as y, convert to numpy array, and obtain all but the last 'x' rows
y = np.array(dF['Prediction'])[:-futureDays]
#print(y)
```

```python
#Split the data to 75% training and 25% testing
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 0.25)
```

```python
#Create the models
#Create Decision tree regressor model
tree = DecisionTreeRegressor().fit(xTrain, yTrain)
#Create the Linear regression model
lR = LinearRegression().fit(xTrain, yTrain)
```

```python
# Obtain the last 'x' rows from the future data set
xFuture = dF.drop(['Prediction'], 1)[:-futureDays]
xFuture = xFuture.tail(futureDays)
xFuture = np.array(xFuture)
```

```python
#Show the tree prediction
treePred = tree.predict(xFuture)
print(treePred)
print()
# Show the  regression model
lrPred = lR.predict(xFuture)
print(lrPred)
```

```python
#Visualize the data
predictions = treePred
valid = dF[x.shape[0]:]
valid['Predictions'] = predictions

#create new figure
plt.figure(figsize=(21,10.5))
plt.title("Bitcoin Prediction Model")
plt.xlabel('Days')
plt.ylabel('$ Price (USD)')
plt.plot(dF['Close'])
plt.plot(valid[['Close', 'Predictions']])
plt.legend(['Orig', 'Val', 'Pred'])
plt.show
```

```python
#Visualize the data
predictions = lrPred
valid = dF[x.shape[0]:]
valid['Predictions'] = predictions

#create new figure
plt.figure(figsize=(21,10.5))
plt.title("Bitcoin Prediction Model")
plt.xlabel('Days')
plt.ylabel('$ Price (USD)')
plt.plot(dF['Close'])
plt.plot(valid[['Close', 'Predictions']])
plt.legend(['Orig', 'Val', 'Pred'])
plt.show
```
